# Adding Columns or Rows

Adding columns or rows can be complicated, but if you do it right, you will find it very handy and make your work more efficient. 

Before adding, you need to pay attention to the original dataframe, especially column names and index. You want the new information match the structure of the original data.

To add rows into exiting dataframe, you can use .append method, and feed a Series in. You can only feed a list into .append, but it won't be a good practice. Here's why

```python
new_data = [1,2,3,4]  # Let's try no index 
stock_data.append(new_data)
stock_data = stock_data.append(new_data) # remember to set value to the original dataframe
stock_data.tail(10)
```

I believe that's not why you want, right. A good practice of adding new data into exiting dataframe should look like this.

```python
new_data = pd.Series([1,2,3,4],name="new",index=["new1","new2","new3","new4"])
stock_data.append(new_data).tail()
```

Alright, how about just adding a empty column? That's simple.

```python
stock_data["new_add"]=np.NaN  #Here you don't need to set value to the original 
stock_data.tail(10)
```

If you want to update the data, you can:

```python
stock_data.loc[0,"new_add"] =1
stock_data.tail(10)
```
