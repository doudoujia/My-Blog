```python
import pandas as pd
import numpy as np
```

Let's get a more complicated dataset to demonstrate this. 

```python
stock_data = pd.read_csv("price.csv")
```

When you deal with stock data, there are a lot of missing information and non-applicable blank cell due to time series calculation.

```python
stock_data[1::10] #show every 10 rows
```

Here's how you remove NaN easily

```python
stock_data.dropna().head()
```

```python
stock_data = stock_data.dropna() #Remember to set value to the original dataframe
```

With drop function, you can also delete a column or row. When you use, drop function, you also need to feed axis parameter to specify column or row, 1 for column, 0 for row.

```python
stock_data.drop("TimeStamp",axis=1).head(2)
```

```python
stock_data.drop(55,axis = 0).head(2)  # Need to feed row index into the function. Here number 55 is the index
```

There's also an easy way to remove duplicated rows

```python
stock_data.drop_duplicates()  #if you don't feed a column name, the function will remove duplicated if it contains in any column
stock_data.drop_duplicates("TimeStamp")[::10] # feed a column name if you have a plan in mind
```

# Special Values

Special Values show up when you perform complex analysis or when you download the raw dataset from the web, including inf, Nan, Null.

You can check them with the following methods

```python
stock_data.isin([np.inf]).head(2) #using np.inf can feed the inf data type value into the function 
```

```python
stock_data.isin(["2017-11-06"]).head(2) #You can also feed in any thing
#but need to feed a list
```

To check NaN/Nan, use...

```python
stock_data.isna().head(2)
```

Use .isnull() to check Null, which shows up very often in web downloaded dataset

```python
stock_data.isnull().head(2)
```

Pay attention when you use them, if you just feed them into index selection, here's what happen.

```python
stock_data[stock_data.isna()].head()  #it return a wired dataframe
```

```python
stock_data.loc[stock_data.isna()] #and this give you an error
```
